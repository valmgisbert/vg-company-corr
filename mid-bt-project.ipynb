```python
## review user score percentage & pos and neg scores
no_count = steam_data[(steam_data['Positive'] == 0) & (steam_data['Negative'] == 0)]
# count = 6828
# catgs = with reviews, no reviews

full_no_count = steam_data[((steam_data['Positive'] == 0) & (steam_data['Negative'] == 0)) & ((steam_data['Recommendations'] == 0) & (steam_data['Average playtime forever'] == 0))]
# count = 6421

no_count_with_recs = steam_data[((steam_data['Positive'] == 0) & (steam_data['Negative'] == 0)) & (steam_data['Recommendations'] != 0)]
# count = 346
# by steam functionality this could not be possible - you need to review to recommend
# we're replacing the positive ones for the user rating
```

```python
## recalculate user score percentage
# step 1: fill empty positives with recs
steam_data['Positive'] = np.where(steam_data['Positive'] == 0, steam_data['Recommendations'], steam_data['Positive'])

# step 2: calculate percentage user score and replace old values
steam_data['User score'] = round(steam_data['Positive']/(steam_data['Positive']+steam_data['Negative'])*100, 2)

# step 3: fill nulls that came up due to 0 values in neg and pos
steam_data['User score'].fillna(0, inplace=True)
```

```python
## add column for community engagement boolean to differentiate between games amongst 0-20k
def fill_community_eng(row):
    if ((row['Positive'] != 0) | (row['Negative'] != 0)) | ((row['Recommendations'] != 0) | (row['Average playtime forever'] != 0)):
        return "Yes"
    else:
        return "No"

steam_data['Community engagement'] = steam_data.apply(fill_community_eng, axis=1)
```

## Correlations

### Main correlations

##### Sales success: estimated owners and revenue range

```python
## correlate between two cols
# crosstab using data (without exploding catgs and genres)
owners_rev_crosstab = pd.crosstab(data['estimated_owners'], data['revenue_range'], margins=False)
owners_rev_crosstab
```

```python
# get a p_value for the crosstab
print('p-value between estimated owners and revnue range is:', st.chi2_contingency(owners_rev_crosstab)[1])
```

Sice our `p_value` is under 0.05, we can differ that there's not enough evidence to reject that the success of a video game can does not depend on the size of a company.

#### Legacy success: estimated owners and avg playtime forever

- H0: a video game that sells a lot is one that is not played often.
- H1: a video game that sells a lot is one that is played often. 

```python
owners_avgpl_crosstab = pd.crosstab(data['estimated_owners'], data['average_playtime_forever'], margins=False)
owners_avgpl_crosstab
```

```python
print('p-value between estimated owners and avg playtime forever is:', st.chi2_contingency(owners_avgpl_crosstab)[1])
```

Since our `p_value` is under 0.05, we can differ that there's not enough evidence to reject that games that are successful in sales are not played often.

```python
# compare against revenue range to check if big company games are played often
avg_rev_crosstab = pd.crosstab(data['average_playtime_forever'], data['revenue_range'], margins=False)
avg_rev_crosstab
```

```python
print('p-value between revenue and avg playtime forever is:', st.chi2_contingency(avg_rev_crosstab)[1])
```

#### Official quality success: revenue range and metacritic score

```python
rev_metacritic_crosstab = pd.crosstab(data['revenue_range'], data['metacritic_score'], margins=False)
rev_metacritic_crosstab
```

```python
print('p-value between revenue range and metacritic score is:', st.chi2_contingency(rev_metacritic_crosstab)[1])
```

Since our `p_value` is under 0.05, we can differ that there's not enough evidence to reject that big companies do not make video games with good official score.

## Conclusions:
- Smaller companies are more present in Steam, taking around 99,4% of the library - while the biggest companies comprise of 0,6% of games
- Sales(est owners): while there are less big-company games on Steam, their audience is bigger in comparison to the indie ones and have more owners.
    - If we compare the numbers between the games with biggest audiences from big companies, we get around 22% of total games
    - while with games from smaller audiences, their bigger audiences are for 1,4% of total games. 
- Avg playtime: consequently, by having both a bigger fanbase and big-scale games, this reflects on the average playtime being bigger for larger-company games.
    - Plus, with the resources at hand, bigger companies make bigger games (hours-wise)
    - while smaller-companies scale down on playtime hours
        - Though that's not to say that indie games are not being played; the biggest avg playtime hours in forever is by a small company visual novel only available in chinese subtitles and japanese audio, with a whooping 90K hours - that's a lot for a game, even a visual novel! 
        - On our biggest company game played the most, we have a popular franchise with 24K hours on average. For this game, these are interestingly low hours - which brings us to the reminder that these big companies don't have just Steam as their sole income; cross-compatibility with consoles is their go-to choice, and this one specifically offers the game on their own website - so in reality, this is just a portion of the hours of their population, and a fact that translates across other big companies.
- Overall comment engagement and presence in steam: though the key part is the community love, and consequently, it's love; comments, reccomendations, scores. We see a constant presence from bigger companies - all of the registered games have at least some community engagement (for better or for worse), and for smaller companies an almost 90% have an online presence! great numbers! 
    - Reviews by community: users aid in bringing notice to the games on both accounts - their score percentages have barely a 10% difference. To get this amount of scores shows these are well-loved games on both accounts.
    - Reviews by metacritic: in more official channels on the other hand, we get more visibility for bigger games (even if their scores are not super high) than indie ones
